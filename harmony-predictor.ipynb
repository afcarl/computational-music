```python
import midi
import numpy as np
import pandas as pd
```

```python
file = 'minuet-in-g.mid'
```

```python
pattern = midi.read_midifile(file)
```

```python
active = np.zeros(88)
time = 0
atime = 0
changes = []
for event in pattern[1]:
    time += event.tick
    if atime != time:
        changes.append({'time': atime, 'resonance': np.array(active)})
        atime = time
    p = False
    pitchOn = False
    if event.name == 'Note On':
        p = True
        pitchOn = True
    elif event.name == 'Note Off':
        p = True
    if p:
        pitch = event.data[0]
        velocity = event.data[1]
        if pitchOn:
            active[pitch] = 1
        else:
            active[pitch] = 0
```

```python
pianoRoll = []
for c in changes:
    pianoRoll.append(c['resonance'])
df = pd.DataFrame(pianoRoll)
```
